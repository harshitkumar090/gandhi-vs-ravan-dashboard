```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib.patches import Circle

# Set style for better visuals
plt.style.use('seaborn-v0_8-darkgrid')

# Sample Zomato data
data = {
    'cuisine_types': ['North Indian', 'Chinese', 'Fast Food', 'South Indian', 'Italian', 'Others'],
    'orders': [450, 380, 520, 290, 210, 350],
    'rating_distribution': ['5 Star', '4 Star', '3 Star', '2 Star', '1 Star'],
    'ratings_count': [1200, 1800, 900, 400, 200],
    'months': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'monthly_orders': [2400, 2650, 2890, 3100, 3450, 3800]
}

# Create figure with subplots
fig = plt.figure(figsize=(16, 10))
fig.suptitle('Zomato Analytics Dashboard', fontsize=20, fontweight='bold', color='#E23744')

# Color schemes
zomato_colors = ['#E23744', '#FF6B6B', '#FFB347', '#87CEEB', '#98D8C8', '#F7B7D2']
rating_colors = ['#1DB954', '#7FCD91', '#FFD93D', '#FF9A3C', '#E23744']

# 1. Pie Chart - Cuisine Distribution
ax1 = plt.subplot(2, 3, 1)
wedges, texts, autotexts = ax1.pie(data['orders'],
                                     labels=data['cuisine_types'],
                                     autopct='%1.1f%%',
                                     colors=zomato_colors,
                                     startangle=90,
                                     explode=[0.05, 0, 0.05, 0, 0, 0],
                                     shadow=True)

for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(9)
    autotext.set_fontweight('bold')

for text in texts:
    text.set_fontsize(10)

ax1.set_title('Orders by Cuisine Type', fontsize=14, fontweight='bold', pad=20)

# 2. Donut Chart - Rating Distribution
ax2 = plt.subplot(2, 3, 2)
wedges, texts, autotexts = ax2.pie(data['ratings_count'],
                                     labels=data['rating_distribution'],
                                     autopct='%1.1f%%',
                                     colors=rating_colors,
                                     startangle=45,
                                     wedgeprops=dict(width=0.5, edgecolor='white'))

# Create center circle for donut
centre_circle = Circle((0, 0), 0.70, fc='white')
ax2.add_artist(centre_circle)

# Add center text
ax2.text(0, 0, f"{sum(data['ratings_count'])}\nTotal\nRatings",
         ha='center', va='center', fontsize=12, fontweight='bold')

for autotext in autotexts:
    autotext.set_color('white')
    autotext.set_fontsize(9)
    autotext.set_fontweight('bold')

for text in texts:
    text.set_fontsize(10)

ax2.set_title('Customer Ratings Distribution', fontsize=14, fontweight='bold', pad=20)

# 3. Column Chart - Monthly Orders
ax3 = plt.subplot(2, 3, 3)
bars = ax3.bar(data['months'], data['monthly_orders'],
               color=zomato_colors[:6],
               edgecolor='black',
               linewidth=1.5,
               alpha=0.8)

# Add value labels on bars
for bar in bars:
    height = bar.get_height()
    ax3.text(bar.get_x() + bar.get_width()/2., height,
             f'{int(height)}',
             ha='center', va='bottom', fontweight='bold', fontsize=10)

ax3.set_xlabel('Month', fontsize=12, fontweight='bold')
ax3.set_ylabel('Number of Orders', fontsize=12, fontweight='bold')
ax3.set_title('Monthly Order Trends', fontsize=14, fontweight='bold', pad=20)
ax3.grid(axis='y', alpha=0.3)
ax3.set_ylim(0, max(data['monthly_orders']) * 1.15)

# 4. Horizontal Bar Chart - Top Cuisines
ax4 = plt.subplot(2, 3, 4)
y_pos = np.arange(len(data['cuisine_types']))
bars = ax4.barh(y_pos, data['orders'], color=zomato_colors, alpha=0.8, edgecolor='black')

ax4.set_yticks(y_pos)
ax4.set_yticklabels(data['cuisine_types'])
ax4.invert_yaxis()
ax4.set_xlabel('Number of Orders', fontsize=12, fontweight='bold')
ax4.set_title('Top Cuisine Performance', fontsize=14, fontweight='bold', pad=20)
ax4.grid(axis='x', alpha=0.3)

# Add value labels
for i, bar in enumerate(bars):
    width = bar.get_width()
    ax4.text(width, bar.get_y() + bar.get_height()/2.,
             f' {int(width)}',
             ha='left', va='center', fontweight='bold', fontsize=10)

# 5. Grouped Column Chart - Comparison
ax5 = plt.subplot(2, 3, 5)
categories = ['Week 1', 'Week 2', 'Week 3', 'Week 4']
delivery = [450, 520, 480, 550]
dine_in = [320, 380, 360, 410]

x = np.arange(len(categories))
width = 0.35

bars1 = ax5.bar(x - width/2, delivery, width, label='Delivery',
                color='#E23744', alpha=0.8, edgecolor='black')
bars2 = ax5.bar(x + width/2, dine_in, width, label='Dine-in',
                color='#FFB347', alpha=0.8, edgecolor='black')

ax5.set_xlabel('Week', fontsize=12, fontweight='bold')
ax5.set_ylabel('Orders', fontsize=12, fontweight='bold')
ax5.set_title('Delivery vs Dine-in Orders', fontsize=14, fontweight='bold', pad=20)
ax5.set_xticks(x)
ax5.set_xticklabels(categories)
ax5.legend(loc='upper left', frameon=True, shadow=True)
ax5.grid(axis='y', alpha=0.3)

# 6. Stacked Column Chart
ax6 = plt.subplot(2, 3, 6)
food_cost = [800, 900, 1000, 1100, 1200, 1300]
delivery_charges = [200, 250, 220, 240, 280, 300]
taxes = [150, 170, 180, 200, 220, 240]

ax6.bar(data['months'], food_cost, label='Food Cost',
        color='#E23744', alpha=0.8, edgecolor='black')
ax6.bar(data['months'], delivery_charges, bottom=food_cost,
        label='Delivery', color='#FFB347', alpha=0.8, edgecolor='black')
ax6.bar(data['months'], taxes,
        bottom=np.array(food_cost) + np.array(delivery_charges),
        label='Taxes', color='#87CEEB', alpha=0.8, edgecolor='black')

ax6.set_xlabel('Month', fontsize=12, fontweight='bold')
ax6.set_ylabel('Revenue (â‚¹1000s)', fontsize=12, fontweight='bold')
ax6.set_title('Revenue Breakdown', fontsize=14, fontweight='bold', pad=20)
ax6.legend(loc='upper left', frameon=True, shadow=True)
ax6.grid(axis='y', alpha=0.3)

# Adjust layout
plt.tight_layout(rect=[0, 0.03, 1, 0.97])
plt.show()

# Print summary statistics
print("\n" + "="*50)
print("ZOMATO ANALYTICS SUMMARY")
print("="*50)
print(f"\nTotal Orders: {sum(data['orders'])}")
print(f"Most Popular Cuisine: {data['cuisine_types'][data['orders'].index(max(data['orders']))]}")
print(f"Total Ratings: {sum(data['ratings_count'])}")
print(f"Average Monthly Orders: {np.mean(data['monthly_orders']):.0f}")
print(f"Growth Rate: {((data['monthly_orders'][-1] - data['monthly_orders'][0]) / data['monthly_orders'][0] * 100):.1f}%")
print("="*50)
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd
import numpy as np

# Funny comparison data
categories = ['Heads Count', 'Peaceful Vibes', 'Fashion Sense', 'Transportation',
              'Diet Preference', 'Anger Management', 'Fan Following', 'Movie Appearances']

gandhi_scores = [1, 100, 30, 20, 90, 40, 50, 45]
ravan_scores = [10, 15, 95, 100, 60, 85, 95, 90]

# Create subplots
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=('ðŸ•Šï¸ Gandhi vs ðŸ‘¿ Ravana: The Ultimate Showdown',
                    'ðŸ“Š Power Stats Comparison',
                    'ðŸŽ­ Historical Impact Timeline',
                    'ðŸ† Who Wins Categories?'),
    specs=[[{'type': 'bar'}, {'type': 'scatter'}],
           [{'type': 'scatter'}, {'type': 'bar'}]],
    vertical_spacing=0.15,
    horizontal_spacing=0.12
)

# 1. Radar-style bar chart (flipped)
fig.add_trace(
    go.Bar(name='Gandhi ðŸ•Šï¸', x=gandhi_scores, y=categories,
           orientation='h', marker=dict(color='#FF9933', line=dict(color='#138808', width=2)),
           text=gandhi_scores, textposition='outside'),
    row=1, col=1
)
fig.add_trace(
    go.Bar(name='Ravana ðŸ‘¿', x=ravan_scores, y=categories,
           orientation='h', marker=dict(color='#8B0000', line=dict(color='#FFD700', width=2)),
           text=ravan_scores, textposition='outside'),
    row=1, col=1
)

# 2. Scatter plot - Battle of Traits
traits = {
    'Wisdom': (95, 85),
    'Strength': (60, 95),
    'Humility': (100, 10),
    'Knowledge': (80, 98),
    'Pride': (5, 100),
    'Compassion': (100, 20)
}

for trait, (g_val, r_val) in traits.items():
    fig.add_trace(
        go.Scatter(x=[g_val], y=[r_val], mode='markers+text',
                   marker=dict(size=20, symbol='star'),
                   text=[trait], textposition='top center',
                   name=trait, showlegend=False),
        row=1, col=2
    )

# Add diagonal line
fig.add_trace(
    go.Scatter(x=[0, 100], y=[0, 100], mode='lines',
               line=dict(dash='dash', color='gray'),
               name='Balance Line', showlegend=False),
    row=1, col=2
)

# 3. Timeline of celebrations
years = np.arange(1947, 2026, 10)
gandhi_followers = [50, 100, 200, 400, 800, 1200, 1800, 2500, 3000]
ravan_effigies = [100, 200, 500, 1000, 2000, 3500, 5000, 7000, 10000]

fig.add_trace(
    go.Scatter(x=years, y=gandhi_followers, mode='lines+markers',
               name='Gandhi Followers (M)', line=dict(color='#FF9933', width=3),
               marker=dict(size=10, symbol='circle')),
    row=2, col=1
)
fig.add_trace(
    go.Scatter(x=years, y=ravan_effigies, mode='lines+markers',
               name='Ravana Effigies Burnt', line=dict(color='#8B0000', width=3),
               marker=dict(size=10, symbol='x')),
    row=2, col=1
)

# 4. Winner count by category
gandhi_wins = sum(1 for g, r in zip(gandhi_scores, ravan_scores) if g > r)
ravan_wins = sum(1 for g, r in zip(gandhi_scores, ravan_scores) if r > g)
ties = sum(1 for g, r in zip(gandhi_scores, ravan_scores) if g == r)

fig.add_trace(
    go.Bar(x=['Gandhi Wins ðŸ•Šï¸', 'Ravana Wins ðŸ‘¿', 'Ties ðŸ¤'],
           y=[gandhi_wins, ravan_wins, ties],
           marker=dict(color=['#FF9933', '#8B0000', '#808080']),
           text=[gandhi_wins, ravan_wins, ties],
           textposition='outside',
           showlegend=False),
    row=2, col=2
)

# Update layout
fig.update_xaxes(title_text="Score (Out of 100)", row=1, col=1)
fig.update_xaxes(title_text="Gandhi's Score â†’", row=1, col=2, range=[0, 105])
fig.update_yaxes(title_text="â† Ravana's Score", row=1, col=2, range=[0, 105])
fig.update_xaxes(title_text="Year", row=2, col=1)
fig.update_yaxes(title_text="Count (Millions/Thousands)", row=2, col=1)
fig.update_xaxes(title_text="Winner", row=2, col=2)
fig.update_yaxes(title_text="Categories Won", row=2, col=2)

fig.update_layout(
    title={
        'text': 'ðŸŽ‰ OCTOBER 2nd SPECIAL: Gandhi Jayanti ðŸ•Šï¸ vs Dussehra ðŸ”¥<br><sub>The Most Peaceful & Most Fiery Day of the Year!</sub>',
        'x': 0.5,
        'xanchor': 'center',
        'font': {'size': 20, 'color': '#2C3E50', 'family': 'Arial Black'}
    },
    height=900,
    showlegend=True,
    legend=dict(x=0.5, y=-0.05, xanchor='center', orientation='h'),
    plot_bgcolor='#FFF8DC',
    paper_bgcolor='#FFFACD',
    font=dict(size=11)
)

# Add funny annotations
fig.add_annotation(
    text="Fun Fact: Gandhi had 1 head but infinite wisdom ðŸ§ <br>Ravana had 10 heads but still made bad choices ðŸ¤¦",
    xref="paper", yref="paper",
    x=0.5, y=-0.12, showarrow=False,
    font=dict(size=12, color='#8B4513'),
    xanchor='center'
)

fig.show()

# Print funny stats
print("=" * 60)
print("ðŸŽŠ OCTOBER 2nd DASHBOARD STATS ðŸŽŠ")
print("=" * 60)
print(f"\nðŸ•Šï¸  Gandhi's Total Score: {sum(gandhi_scores)}")
print(f"ðŸ‘¿ Ravana's Total Score: {sum(ravan_scores)}")
print(f"\nðŸ† Gandhi wins in {gandhi_wins} categories!")
print(f"ðŸ”¥ Ravana wins in {ravan_wins} categories!")
print(f"\nðŸ’¡ Lesson: More heads â‰  Better decisions! ðŸ˜„")
print("=" * 60)
```
